```python
import requests
from bs4 import BeautifulSoup 
import csv 
```

```python

BASE_URL = "https://economictimes.indiatimes.com"
RECO_URL = BASE_URL + "/markets/stocks/recos?from=mdr"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}

response = requests.get(RECO_URL, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
print(response.content)
```

```python
# Find all the stories
stories = soup.find_all("div", class_="eachStory")
```

```python
stories
```

```python
for story in stories:
    try:
        # Extract title and article link
        h3_tag = story.find("h3")
        a_tag = h3_tag.find("a") if h3_tag else None
        title = a_tag.get_text(strip=True) if a_tag else "N/A"
        url = BASE_URL + a_tag['href'] if a_tag and a_tag.get('href') else "N/A"

        # Extract date and time
        time_tag = story.find("time", class_="date-format")
        datetime = time_tag.get_text(strip=True) if time_tag else "N/A"

        # Extract summary paragraph
        p_tag = story.find("p")
        summary = p_tag.get_text(strip=True) if p_tag else "N/A"

        # Extract stock price and change (optional)
        stock_widget = story.find("div", class_="stockDataWidget")
        price = stock_widget.find("span", class_="stock_price").get_text(strip=True) if stock_widget else "N/A"
        change = stock_widget.find("span", class_="stock_change").get_text(strip=True) if stock_widget else "N/A"

        print(f"Title     : {title}")
        print(f"Date/Time : {datetime}")
        print(f"Summary   : {summary}")
        print(f"URL       : {url}")
        print(f"Price     : {price}")
        print(f"Change    : {change}")
        print("-" * 80)

    except Exception as e:
        print(f"Error parsing story: {e}")
```

Load More as well 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import json 

all_data = [] 

# Setup headless browser
options = Options()
options.headless = True
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

BASE_URL = "https://economictimes.indiatimes.com"
RECO_URL = BASE_URL + "/markets/stocks/recos?from=mdr"
driver.get(RECO_URL)

from selenium.common.exceptions import ElementClickInterceptedException, NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Click the 'Load More' button 60 times
for i in range(60):
    try:
        print(f"Clicking Load More... ({i+1}/60)")
        load_more = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "autoload_continue"))
        )

        # Scroll to the button and click using JS
        driver.execute_script("arguments[0].scrollIntoView(true);", load_more)
        time.sleep(0.5)

        driver.execute_script("arguments[0].click();", load_more)
        time.sleep(2.5)

    except (ElementClickInterceptedException, NoSuchElementException) as e:
        print(f"[{i}] Load more button not clickable or exhausted.")
        break
    except Exception as e:
        print(f"[{i}] Unexpected error: {e}")
        break


# Get the final page source after loading
soup = BeautifulSoup(driver.page_source, 'html.parser')
driver.quit()

# Parse all story blocks
stories = soup.find_all("div", class_="eachStory")

for story in stories:
    try:
        # Title and URL
        a_tag = story.find("h3").find("a")
        title = a_tag.text.strip() if a_tag else "N/A"
        url = BASE_URL + a_tag['href'] if a_tag else "N/A"

        # Date
        time_tag = story.find("time", class_="date-format")
        date = time_tag.text.strip() if time_tag else "N/A"

        # Summary
        p_tag = story.find("p")
        summary = p_tag.text.strip() if p_tag else "N/A"

        # Stock price and change
        stock_widget = story.find("div", class_="stockDataWidget")
        price = stock_widget.find("span", class_="stock_price").text.strip() if stock_widget else "N/A"
        change = stock_widget.find("span", class_="stock_change").text.strip() if stock_widget else "N/A"

        all_data.append({
            "title": title,
            "date": date,
            "summary": summary,
            "url": url,
            "price": price,
            "change": change
        })

    except Exception as e:
        print("Error parsing story:", e)
```

```python
with open("26_sept_stock_recos.json", "w", encoding="utf-8") as f:
    json.dump(all_data, f, indent=2, ensure_ascii=False)
```
